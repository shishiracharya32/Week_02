GroupBy Objects in Panda

```python
movie_dataset.columns
```

```python
movie_dataset['genre'].value_counts()
```

```python
genre=movie_dataset.groupby('genre')
genre 
```

```python
#find the genre with highest avg score
genre['IMDB Score'].mean().sort_values(ascending=False).head(1) #First we grouped genre and checked the imdb score of the particular group and we checked the mean of each genre
```

```python
#find the director with most numbers of vote
movie_dataset.columns
```

```python
director=movie_dataset.groupby('director')
```

```python
director['votes'].sum().sort_values(ascending=False).head()
```

```python
genre.value_counts() #return the number of groups in objects
```

```python
genre.size() #return the number of data in each group
```

```python
#First and last item of group by
genre.first() #first item of each group
genre.last() #last item of each group
genre.nth(5).set_index('genre')
```

```python
#get_group
genre.get_group('Romance')
```

```python
genre.get_group('Sci-Fi')['Movies_Name']
```

```python
#agg method multiple

genre[['IMDB Score','gross']].agg(['min','median','max','sum','count'])
```

```python
genre.agg({'IMDB Score':['mean','median'],
           'gross':'sum'})
```

```python
#string
country=["Nepal","USA","Canada","Russia","Ukraine","Pakistan"]
country
```

```python
pd.Series(country) #Creating a series for 1D array only.
```

```python
population=[3,150,50,10,5,50]
```

```python
pd.Series(population) #series for integer data type
```

```python
#custom indexing
marks=(90,50,40,70,80)
sub=("English","Nepali","Math","Science","Computer")
pd.Series(marks,index=sub) #index='' gives custom index

```

```python
#setting a name
#custom indexing
marks=(90,50,40,70,80)
sub=("English","Nepali","Math","Science","Computer")
pd.Series(marks,index=sub,name="Shishir Marks") #name="Series name"

```

Series Math Method

```python
#size
print(marks)
print(marks.size) #Count null value also
```

```python
#count
marks.count() # do not count null value
```

```python
##sum,mean,median,std,mode
print(marks.sum())
print(marks.mean())
print(marks.mode())
print(marks.std())
print(marks.median())
```

```python
#min/max
print(marks.min())
print(marks.max())
```

```python
#describe
marks.describe() #display the statistical summary of the data
```
